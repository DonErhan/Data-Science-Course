```python
# Make blobs
# Consist from 500 samples, 3 centers, 6 features, cluster std is 5 and random state 420

X, y = make_blobs(n_samples=500, n_features=6, centers=3, cluster_std=5,random_state=420)
```

```python
# Linear Discriminant Analizer, 
# Reduce dimensionality from 6 features to 2 features

lda = LinearDiscriminantAnalysis(n_components=2)
lda.fit(X, y)
X_transformed = lda.transform(X)
```

```python
# Split to train and test

X_train, X_test, y_train, y_test = train_test_split(X_transformed, y)
```

```python
# Accuracy classification score

accuracy_score(y_pred, y_test)
```
