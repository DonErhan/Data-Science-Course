# Best feature selection with Kydavra based on correlation

### Pearson Correlation Selector

```python
# Pearson correlation selector

P_selector = PearsonCorrelationSelector(min_corr = 0.2)
P_columns = P_selector.select(df, 'Outcome')
```

```python
# Pearson correlation selector selected columns

df[P_columns]
```

### Kendal Correlation Selector

```python
# Kendal Correlation Selector

K_selector = KendallCorrelationSelector(min_corr = 0.2)
K_columns = K_selector.select(df, 'Outcome')
```

```python
# Kendal correlation selector selected columns

df[K_columns]
```

### Spearman Correlation Selector

```python
# Spearman Correlation Selector

S_selector = SpearmanCorrelationSelector(min_corr = 0.2)
S_columns = S_selector.select(df,'Outcome')
```

```python
# Spearman correlation selector selected columns

df[S_columns]
```

### Point Biserial Correlation Selector

```python
# Point Biserial Corr Selector

PB_selector = PointBiserialCorrSelector()
PB_columns = PB_selector.select(df, 'Outcome')
```

```python
# Point Biserial Correlation Selector selected columns

df[PB_columns]
```
